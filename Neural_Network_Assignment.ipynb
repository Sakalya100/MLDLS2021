```python
#importing the required libraries
import tensorflow as tf
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
#import Dataset
dataset = pd.read_csv('/content/sample_data/Car_price.csv')
print(dataset)
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
xs=np.array(X,dtype=float)
ys=np.array(Y,dtype=float)
```

```python
x=np.array([],dtype=float)
y=np.array([2,4,6,8],dtype=float)
```

```python
basic= tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
basic.compile(optimizer='sgd',loss='mean_squared_error')
basic.fit(X,Y,epochs=500)
```

```python
model= tf.keras.Sequential([keras.layers.Dense(units=3,activation='relu',input_shape=[1])])
model.add(Dense(1))
print(model.summary())
```

```python
tf.keras.utils.plot_model(
    model,
    to_file='model.png',
    show_shapes=True,
    show_dtype=False,
    show_layer_names=True,
    rankdir='TB',
    expand_nested=False,
    dpi=96,
)
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
run_model=model.fit(X,Y,epochs=700)
```

```python
print(model.predict([145.0]))
```

```python
print(run_model.history.keys())
```

```python
plt.plot(run_model.history['loss'])
plt.title('Model Loss')
plt.xlabel('loss')
plt.ylabel('epochs')
plt.show()
```
